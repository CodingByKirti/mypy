```python
wb = openpyxl.load_workbook("xlSample.xlsx")
ws = wb.active
ws
```

```python
ws.title ="Sales"
ws.sheet_properties.tabColor ="222222"
# ws.title
wb.save("xlSample.xlsx")
```

```python
## add data into worksheet
list_names=['James','Williams','Johans','Lila','Samantha']

for i in range(0, len(list_names)):
    ws["A"+str(i+1)].value = list_names[i]
```

```python
# another way
for i in range(0, len(list_names)):
    ws.cell(row = i+1, column =2).value = list_names[i]
wb.save("xlSample.xlsx")
```

```python
# read data from speadsheet

for col in ws['A']:
    print(col.value)
```

```python
for col in ws['B']:
    print(col.value)
```

```python
for col in ws["A":"B"]:
    print(col)
```

```python
# following will show sheetname along with cell number
# for row in ws.iter_rows(min_row=1,max_row=5, min_col=1, max_col=2):
#     print(row)


# to get data : add one parameter : values_only = True
for row in ws.iter_rows(min_row=1,max_row=5, min_col=1, max_col=2, values_only=True):
    print(row)
```

```python
# read in form of col
for col in ws.iter_cols(min_row=1,max_row=5, min_col=1, max_col=2, values_only=True):
    print(col)
```
