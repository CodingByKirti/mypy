```python
import os
import win32com.client as win32

def set_active_cell_b3(folder_path):
    excel = win32.Dispatch("Excel.Application")
    excel.Visible = False
    excel.DisplayAlerts = False

    # Loop through all Excel files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith((".xlsx", ".xlsm", ".xls")):  # Excel file types
            file_path = os.path.join(folder_path, filename)
            print(f"Processing: {file_path}")

            try:
                wb = excel.Workbooks.Open(file_path)

                # Loop through each sheet and select B3
                for ws in wb.Sheets:
                    ws.Activate()
                    ws.Range("B3").Select()

                # Save changes and close
                wb.Save()
                wb.Close(SaveChanges=True)

            except Exception as e:
                print(f"Error processing {filename}: {e}")

    excel.Quit()

if __name__ == "__main__":
    folder_path = r"C:\Users\YourName\Documents\ExcelFiles"  # change your path
    set_active_cell_b3(folder_path)
```

```python
import os
import win32com.client as win32
from tkinter import Tk, filedialog

def paste_values_and_cleanup_save_as():
    # Ask user to select folder
    Tk().withdraw()
    source_folder = filedialog.askdirectory(title="Select Folder Containing Excel Files")

    if not source_folder:
        print("No folder selected. Exiting.")
        return

    # Create output folder
    parent_dir = os.path.dirname(source_folder)
    folder_name = os.path.basename(source_folder)
    output_folder = os.path.join(parent_dir, f"{folder_name}_PV")
    os.makedirs(output_folder, exist_ok=True)

    # Launch Excel
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    excel.Visible = False
    excel.DisplayAlerts = False  # Suppress prompts

    for filename in os.listdir(source_folder):
        if filename.endswith('.xlsx'):
            source_path = os.path.join(source_folder, filename)
            name_part = os.path.splitext(filename)[0]
            output_filename = f"{name_part}_PV.xlsx"
            output_path = os.path.join(output_folder, output_filename)

            print(f"Processing: {filename}")

            wb = excel.Workbooks.Open(source_path)

            # Paste values safely and set active cell without using Select
            for ws in wb.Worksheets:
                ws.UsedRange.Value = ws.UsedRange.Value
                ws.ScrollRow = 3       # Ensures row 3 is visible when opening
                ws.ScrollColumn = 2    # Ensures column B is visible
                ws.Range("B3").Activate()  # Sets B3 as the active cell without Select

            # Delete unwanted sheets
            for sheet_name in ['Data', 'Date']:
                try:
                    wb.Worksheets(sheet_name).Delete()
                except Exception:
                    pass

            # Make first sheet active
            wb.Sheets(1).Activate()

            # Clear any selection / copy mode
            excel.CutCopyMode = False

            # Save copy to output folder
            wb.SaveCopyAs(output_path)
            wb.Close(SaveChanges=False)

    excel.Quit()
    print(f"✅ All processed files saved to: {output_folder}")

if __name__ == "__main__":
    paste_values_and_cleanup_save_as()
```

```python
import os
import win32com.client as win32

def paste_values_and_cleanup(folder_path):
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    excel.Visible = False

    for filename in os.listdir(folder_path):
        if filename.endswith('.xlsx'):
            file_path = os.path.join(folder_path, filename)
            print(f"Processing: {filename}")

            wb = excel.Workbooks.Open(file_path)

            # Paste values for all sheets
            for ws in wb.Worksheets:
                ws.Cells.Copy()
                ws.Cells.PasteSpecial(Paste=-4163)  # xlPasteValues

            # Delete 'Data' and 'Date' sheets if they exist
            for sheet_name in ['Online', 'Offline']:
                try:
                    wb.Worksheets(sheet_name).Delete()
                    print(f"Deleted sheet: {sheet_name}")
                except Exception:
                    pass  # Sheet does not exist — skip

            # Set first sheet as active and cell B3 selected
            if wb.Sheets.Count > 0:
                first_sheet = wb.Sheets(1)
                first_sheet.Activate()
                first_sheet.Range("B3").Select()

            # Save and close
            wb.Save()
            wb.Close(SaveChanges=True)

    excel.Quit()
    print("Done!")
    
folder_path = r"C:\\Users\\KS\\K"
paste_values_and_cleanup(folder_path)
```

```python
import os
import win32com.client as win32
from tkinter import Tk, filedialog

def paste_values_and_cleanup_save_as():
    # Ask user to select folder
    Tk().withdraw()
    source_folder = filedialog.askdirectory(title="Select Folder Containing Excel Files")

    if not source_folder:
        print("No folder selected. Exiting.")
        return

    # Create output folder
    parent_dir = os.path.dirname(source_folder)
    folder_name = os.path.basename(source_folder)
    output_folder = os.path.join(parent_dir, f"{folder_name}_PV")
    os.makedirs(output_folder, exist_ok=True)

    # Clean list of only actual Excel files (not temp)
    file_list = [
        f for f in os.listdir(source_folder)
        if f.endswith(".xlsx") and not f.startswith("~$")
    ]

    if not file_list:
        print("No valid Excel files found in the folder.")
        return

    # Launch Excel
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    excel.Visible = False
    excel.DisplayAlerts = False  # Suppress delete prompts

    for filename in file_list:
        source_path = os.path.join(source_folder, filename)
        name_part = os.path.splitext(filename)[0]
        output_filename = f"{name_part}_PV.xlsx"
        output_path = os.path.join(output_folder, output_filename)

        print(f"Processing: {filename}")

        try:
            wb = excel.Workbooks.Open(source_path)

            # Paste values
            for ws in wb.Worksheets:
                ws.Cells.Copy()
                ws.Cells.PasteSpecial(Paste=-4163)  # xlPasteValues

            # Delete specified sheets if they exist
            for sheet_name in ['Online ', 'Date']:
                try:
                    wb.Worksheets(sheet_name).Delete()
                except Exception:
                    pass

            # Set first sheet and B3
            if wb.Sheets.Count > 0:
                first_sheet = wb.Sheets(1)
                first_sheet.Activate()
                first_sheet.Range("B3").Select()

            wb.SaveCopyAs(output_path)
            wb.Close(SaveChanges=False)

        except Exception as e:
            print(f"❌ Failed to process {filename}: {e}")

    excel.Quit()
    print(f"\n✅ All valid files saved to: {output_folder}")

# Run it
paste_values_and_cleanup_save_as()
```
