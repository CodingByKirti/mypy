```python
import asyncio
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeoutError

async def run():
    chrome_path = "C:/Program Files/Google/Chrome/Application/chrome.exe"  # Adjust if necessary

    async with async_playwright() as playwright:
        # Launch Chrome browser using the Chromium engine
        browser = await playwright.chromium.launch(
            executable_path=chrome_path, 
            headless=False  # Run in a full browser window, not headless
        )
        context = await browser.new_context()
        page = await context.new_page()

        # Navigate to the initial website
        await page.goto("https://www.example.com")  # Replace with actual URL

        # Wait for the page to fully load
        await page.wait_for_load_state('networkidle')

        # Click the SSO button
        button_text_selector = "//button[text()=\"My organization's single sign-on (SSO)\"]"
        try:
            await page.wait_for_selector(f'xpath={button_text_selector}', timeout=10000)
            await page.click(f'xpath={button_text_selector}')
            print("SSO button clicked successfully!")
        except PlaywrightTimeoutError:
            print("Could not find the SSO button on the page.")

        # Wait for the dashboard to load
        await page.wait_for_load_state('networkidle')

        print("Dashboard loaded. Interacting with filters...")

        # Select first filter ("Finance" option)
        try:
            await page.wait_for_selector('select#firstFilter', timeout=10000)  # Replace with actual ID
            await page.select_option('select#firstFilter', label='Finance')  # Use 'label' or 'value'
            print("Selected 'Finance' in the first filter.")
        except Exception as e:
            print(f"Error selecting first filter: {e}")

        # Loop through all options in the second filter
        try:
            await page.wait_for_selector('select#secondFilter', timeout=10000)  # Replace with actual ID
            second_filter_options = await page.query_selector_all('select#secondFilter option')

            for option in second_filter_options:
                option_value = await option.get_attribute('value')
                option_text = await option.inner_text()
                
                print(f"Selecting second filter option: {option_text}")
                await page.select_option('select#secondFilter', value=option_value)

                # Click ellipses to download report
                await page.click('.ellipses-menu')  # Replace with actual selector for ellipses
                await page.click('button#downloadReport')  # Replace with actual selector for download button

                print(f"Report for {option_text} downloaded.")
                await asyncio.sleep(2)  # Delay between downloads
        except Exception as e:
            print(f"Error processing second filter options: {e}")

        # Keep the browser open for inspection
        await asyncio.sleep(300)  # Keep open for 5 minutes or until manually closed

if __name__ == "__main__":
    asyncio.run(run())
```
